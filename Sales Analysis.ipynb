# Exploratory Data Analysis (EDA)



```python
df.head()
```

####  1. Average Amount Spent by Age

```python

plt.figure(figsize=(12, 6))
age_amount = df.groupby('Age')['Amount'].mean().sort_index()
plt.plot(age_amount.index, age_amount.values, marker='o', linewidth=2, markersize=6)
plt.title('Average Purchase Amount by Customer Age')
plt.xlabel('Age')
plt.ylabel('Average Amount Spent')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

####  2. Product Category Performance

```python
plt.figure(figsize=(12, 6))
category_metrics = df.groupby('Product_Category').agg({
    'Orders': 'mean',
    'Amount': 'mean'
}).sort_values('Amount')

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Product Category')
ax1.set_ylabel('Average Orders', color='tab:blue')
line1 = ax1.plot(category_metrics.index, 
                 category_metrics['Orders'],
                 color='tab:blue', 
                 marker='o', 
                 label='Avg Orders')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Average Amount', color='tab:red')
line2 = ax2.plot(category_metrics.index, 
                 category_metrics['Amount'],
                 color='tab:red', 
                 marker='s', 
                 label='Avg Amount')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Product Category Performance Analysis')
plt.xticks(rotation=45)

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

plt.tight_layout()
plt.show()

```

####  3. Marital Status Impact Over Age Groups

```python
plt.figure(figsize=(12, 6))
marital_age = df.pivot_table(
    index='Age Group',
    columns='Marital_Status',
    values='Amount',
    aggfunc='mean'
).sort_index()

for status in marital_age.columns:
    plt.plot(range(len(marital_age.index)),
            marital_age[status],
            marker='o',
            label=status,
            linewidth=2)

plt.xticks(range(len(marital_age.index)), marital_age.index, rotation=45)
plt.title('Purchase Amount Trends by Marital Status Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Average Purchase Amount')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

####  4. Histogram (Matplotlib) - Age Distribution

```python

plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=30, edgecolor='black')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

####  5. Pie Chart (Matplotlib) - Product Category Distribution

```python

plt.figure(figsize=(10, 10))
product_counts = df['Product_Category'].value_counts()
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%')
plt.title('Product Category Distribution')
plt.show()
```

####  6. Area Plot (Matplotlib) - Orders by Zone and Gender

```python
pivot_data = df.pivot_table(index='Zone', columns='Gender', values='Orders', aggfunc='sum')
pivot_data.plot(kind='area', stacked=True, figsize=(10, 6))
plt.title('Orders by Zone and Gender')
plt.xlabel('Zone')
plt.ylabel('Total Orders')
plt.show()
```

#### 7. Heatmap (Seaborn) - Correlation Matrix

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=[np.number]).columns
correlation = df[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='flag', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

#### 8. Pair Plot (Seaborn) - Relationships between numeric variables

```python

sns.pairplot(df[['Age', 'Orders', 'Amount', 'Gender']], hue='Gender')
plt.suptitle('Pair Plot of Key Variables', y=1.02)
plt.show()
```

#### 9. Joint Plot (Seaborn) - Orders vs Amount

```python
sns.jointplot(data=df, x='Orders', y='Amount', kind='reg', height=8)
plt.suptitle('Joint Distribution of Orders and Amount', y=1.02)
plt.show()
```

#### 10. Violin Plot (Seaborn) - Amount Distribution by Zone

```python

plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Zone', y='Amount', hue='Gender')
plt.title('Amount Distribution by Zone and Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. KDE Plot (Seaborn) - Age Distribution by Marital Status
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Age', hue='Marital_Status', fill=True)
plt.title('Age Distribution by Marital Status')
plt.show()
```
